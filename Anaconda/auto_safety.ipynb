```python
from selenium import webdriver
import time
```

```python
# Flash 사용 설정
prefs = {
    "profile.default_content_setting_values.plugins": 1,
    "profile.content_settings.plugin_whitelist.adobe-flash-player": 1,
    "profile.content_settings.exceptions.plugins.*,*.per_resource.adobe-flash-player": 1,
    "PluginsAllowedForUrls": "https://url.com"
}


options = webdriver.ChromeOptions()
options.add_experimental_option("prefs",prefs)

driver = webdriver.Chrome(r'C:\python\Anaconda\chromedriver.exe', chrome_options=options) # chromedriver.exe 경로 설정
```

```python
driver.get("http://safety.kw.ac.kr")
time.sleep(0.5)

login_id = driver.find_element_by_name("UniqueKey");
login_id.send_keys("2016112503");       # 학번 입력
login_name = driver.find_element_by_name("UserName");
login_name.send_keys("김경도");         # 이름 입력

time.sleep(0.3)
driver.find_element_by_id("btnStudent").click()

time.sleep(0.1)
driver.get("http://safety.kw.ac.kr/Edu/OnLineEdu")
```

```python
# 수강할 과정명 선택 후 실행

for i in range(0, 13):   # 듣고자하는 강의 수 만큼 범위 설정     ex) 6개 => range(0, 6)
    driver.find_element_by_class_name("courseBtn").click()
    driver.switch_to_window(driver.window_handles[1])
    current_url = driver.current_url + "&currentPage=1"
    driver.get(current_url)
    
    for j in range(1, 30) :
        time.sleep(5.5)
        driver.execute_script("page_complete();")
        try :
            alert = driver.switch_to_alert()
            alert.accept()
            break
        except :
            print(j)
        driver.execute_script("next_page();")
    driver.close()
    driver.switch_to_window(driver.window_handles[0])
    time.sleep(0.5)
```
