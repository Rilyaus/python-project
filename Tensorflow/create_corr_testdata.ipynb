```python
import numpy as np
import pandas as pd
import xlsxwriter
```

```python
clist = ['01', '02', '03', '04', '05', '06', '07', '08']

for cNum in clist :
    data = open('./Data/ParseNew/Test_case' + cNum + '.dat')
    corr_wind_u = open('./Result/d2_corrected_wind/wind_u/correction_wind_u_' + cNum + '.dat')
    corr_wind_v = open('./Result/d2_corrected_wind/wind_v/correction_wind_v_' + cNum + '.dat')

    read_data = np.loadtxt(data)
    read_wind_u = np.loadtxt(corr_wind_u)
    read_wind_v = np.loadtxt(corr_wind_v)

    data.close()
    corr_wind_u.close()
    corr_wind_v.close()

    float_formatter = lambda x: "%.4f" % x
    np.set_printoptions(formatter={'float_kind':float_formatter})

    for i in range(0, len(read_data)) :
        read_data[i, 4] = read_wind_u[i]
        read_data[i, 5] = read_wind_v[i]

    np.savetxt('corrected_case' + cNum + '.dat', read_data, delimiter=" ", fmt="%.4f")
```
